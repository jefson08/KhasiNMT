#7. Import Necessary Package for Plotting and Finding Cosine Similarity



```python
import torch
import dill as pickle
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances
import numpy as np
import heapq
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
import random
```

# 8. Load Vocabulary and Embedding Weights

```python
SRC = pickle.load(open(f'weights/SRC.pkl', 'rb'))
embedding = torch.load(f'weights/embedding_weights.pt')
```

# 9. Similarity Function and Plotting Function



```python
def find_similar(similar, no_similar):
	near= [cosine_similarity(similar.reshape(1,-1),embedding[x].reshape(1,-1)) for x in range(embedding.shape[0])]
	#print(near[0:10])
	near= np.array(near).reshape(-1,1)
	nearest=heapq.nlargest(no_similar, range(len(near)), near.take)
	largests = heapq.nlargest(no_similar, near)
	return nearest, largests

def plotting(combine):
  #plt.figure(figsize=(8,6))
  df=pd.DataFrame(combine['data'])
  scaler=StandardScaler()
  scaler.fit(df)
  scaled_data = scaler.transform(df)
  pca=PCA(n_components=2)
  pca.fit(scaled_data)
  x_pca=pca.transform(scaled_data)
  sns.set()
  plt.xlim(x_pca[:,0].min()*1.5, x_pca[:,0].max()*1.5)
  plt.ylim(x_pca[:,1].min()*1.5, x_pca[:,1].max()*1.5)
  plt.scatter(x_pca[:,0], x_pca[:,1], cmap='plasma', color='r')
  plt.xlabel('First principal compoent')
  plt.ylabel('Second principal component')
  #plt.xticks(rotation=45)
  #plt.yticks(rotation=90)
  for i in range(len(combine['words'])):
      col = (np.random.random(), np.random.random(), np.random.random())
      plt.annotate(combine['words'][i], (x_pca[i,0], x_pca[i,1] + 0.2), xytext=(x_pca[i,0]+1, x_pca[i,1] + 1))
      plt.quiver(0, 0, x_pca[i,0], x_pca[i,1], scale_units='xy', angles='xy', color=col, scale=1) 

def plotting_tsne(combine):
    # plt.figure(figsize=(8,6))
    df = pd.DataFrame(combine['data'])
    scaler = StandardScaler()
    scaler.fit(df)
    scaled_data = scaler.transform(df)
    # pca
    #pca = PCA(n_components=2)
    # find tsne coords for 2 dimensions
    tsne = TSNE(n_components=2, random_state=0)
    # pca.fit(scaled_data)
    #x_pca = pca.transform(scaled_data)
    x_pca = tsne.fit_transform(scaled_data)
    sns.set()
    plt.xlim(x_pca[:, 0].min()*1.5, x_pca[:, 0].max()*1.5)
    plt.ylim(x_pca[:, 1].min()*1.5, x_pca[:, 1].max()*1.5)
    plt.scatter(x_pca[:, 0], x_pca[:, 1], cmap='plasma', color='r')
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    # plt.xticks(rotation=45)
    # plt.yticks(rotation=90)
    for i in range(len(combine['words'])):
        col = (np.random.random(), np.random.random(), np.random.random())
        plt.annotate(combine['words'][i], (x_pca[i, 0], x_pca[i,
                     1] + 0.2), xytext=(x_pca[i, 0]+1, x_pca[i, 1] + 1))
        # plt.quiver(0, 0, x_pca[i, 0], x_pca[i, 1],
        #            scale_units='xy', angles='xy', color=col, scale=1)

    plt.show()

def plotting_3d(combine):
    # plt.figure(figsize=(8,6))
    df = pd.DataFrame(combine['data'])
    scaler = StandardScaler()
    scaler.fit(df)
    scaled_data = scaler.transform(df)
    # pca = PCA(n_components=3)
    # pca.fit(scaled_data)
    # x_pca = pca.transform(scaled_data)
    tsne = TSNE(n_components=3, random_state=0)
    x_pca = tsne.fit_transform(scaled_data)
    sns.set()
    fig = plt.figure()
    ax = plt.axes(projection='3d')

    ax.set_xlim(x_pca[:, 0].min()*1.5, x_pca[:, 0].max()*1.5)
    ax.set_ylim(x_pca[:, 1].min()*1.5, x_pca[:, 1].max()*1.5)
    ax.set_zlim(x_pca[:, 2].min()*1.5, x_pca[:, 2].max()*1.5)
    ax.view_init(25, 65)
    start = [0, 0, 0]
    for i, _ in enumerate(x_pca):
        ax.quiver(start[0], start[1], start[2],
                  x_pca[i, 0], x_pca[i, 1], x_pca[i, 2], color=(np.random.random(), np.random.random(), np.random.random()), alpha=.8, lw=2, arrow_length_ratio=0.1)

    for i in range(len(combine['words'])):
        # col = (np.random.random(), np.random.random(), np.random.random())
        ax.text(x_pca[i, 0], x_pca[i, 1], x_pca[i, 2], combine['words'][i])
    plt.show()
```
